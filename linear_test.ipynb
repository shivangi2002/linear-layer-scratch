# Dataset

```python
dataset = [
    ([1,3],4),
    ([3,6],9),
    ([5,7],12)
]
```

```python
for inputs, target in dataset:
    print(inputs, target)
```

```python
layer = Linear(input_size=2)
loss_history = []
for step in range(60):
    loss_sum = 0
    for inputs, target in dataset:
        out = layer.forward(inputs)
        print(out)
        loss_sum += loss(out, target)
        layer.backward(target)
    loss_history.append(loss_sum)
    print("loss progress:", loss_history)  
    
```

```python
layer = Linear(input_size=2)
loss_history = []
for step in range(60):
    loss_sum = 0
    for inputs, target in dataset:
        out = layer.forward(inputs)
        loss_sum += loss(out, target)
        layer.backward(target)
    
    average_loss = loss_sum / len(dataset)
    print("epoch:", step, "loss:", average_loss, "lr:", layer.lr)
    loss_history.append(average_loss)
print("loss progress:", loss_history)  
    
```
